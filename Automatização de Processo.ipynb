```python
import pyautogui 
import pyperclip 
import time 
pyautogui.PAUSE = 1

#1 - Entrar no sistema (entrar no link)
pyautogui.hotkey ("ctrl", "t")
link =  "https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing"
pyperclip.copy(link) 
pyautogui.hotkey("ctrl", "v")
pyautogui.press ("enter")

#2 - Navegar no sistema (entrar na pasta Exportar)
time.sleep(5)
pyautogui.click(x=427, y=264, clicks=2)
#3 - Baixar o arquivo de vendas 
pyautogui.click(x=398, y=337)
pyautogui.click(x=1710, y=158)
pyautogui.click(x=1586, y=554)
time.sleep(4)
pyautogui.press("enter")
```

```python
#4 - Calcular faturamento e quantidade de produtos vendidos
import pandas as pd
tabela = pd.read_excel(r"E:\Vendas - Dez.xlsx")
display(tabela)
faturamento = tabela["Valor Final"].sum()
quantidade = tabela["Quantidade"].sum() 
```

```python
#5 - Enviar email 
pyautogui.hotkey ("ctrl", "t")
link = "https://tela inicial do seu email"
pyperclip.copy(link) 
pyautogui.hotkey("ctrl", "v")
pyautogui.press ("enter")


#clica no botão escrever
time.sleep(5)
pyautogui.click(x=65, y=166)

#escrever pra quem vou mandar
pyautogui.write("destinatário@gmail.com")
pyautogui.press("tab") #escolher o email
pyautogui.press("tab") #passar pro campo de assunto

#escreve o assunto 
assunto = "Relatório de Vendas"
pyperclip.copy(assunto)
pyautogui.hotkey("ctrl", "v")
pyautogui.press("tab") #passar pro campo de email

#escreve o email
texto = f""" 
Prezados, bom dia

O faturamento foi de R${faturamento:,.2f} 
A quantidade de produtos vendidos foi de {quantidade:,}

Abs
Atenciosamente, Samuel

"""
pyautogui.write(texto)
#envia o email 
pyautogui.hotkey("ctrl", "enter")

```
